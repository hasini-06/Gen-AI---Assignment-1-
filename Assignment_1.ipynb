```python
from transformers import pipeline

text = pipeline("text-generation", model="gpt2")

prompt = "Write a simple English poem about the moon."

print(text(prompt))
```

```python
from transformers import pipeline

text = pipeline("text-generation", model="EleutherAI/gpt-neo-125M")
prompt = "Explain about photosynthesis in simple terms"
print(text(prompt))   #text-generation
```

```python
from transformers import pipeline

text = pipeline("summarization", model="Falconsai/text_summarization")
prompt="""Artificial Intelligence (AI) has become a transformative force in modern society, revolutionizing the way we live and work.
From virtual assistants like Siri and Alexa to advanced machine learning algorithms used in healthcare and finance, AI is deeply integrated into daily life.
Its ability to process massive amounts of data and learn from it enables automation of complex tasks, improves decision-making, and enhances user experiences.
However, alongside its benefits, AI also raises concerns about privacy, job displacement, and ethical responsibilities."""

print(text(prompt))   #summarization
```

```python
from transformers import pipeline

text = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")
context="""Dehydration occurs when your body loses more fluids than it takes in.
This imbalance can interfere with normal body functions.
Common causes include excessive sweating, vomiting, diarrhea, or not drinking enough water.
Symptoms of dehydration include dry mouth, fatigue, dizziness, and dark-colored urine.
"""
question = "What are the symptoms of dehydration?"
print(text(question=question,context=context))    #questioning-answering
```

```python
from transformers import pipeline

translate = pipeline("translation", model="Helsinki-NLP/opus-mt-en-el")

prompt = "How are you?"

print(translate(prompt))    #translation
```

```python
from transformers import pipeline

generator = pipeline("text2text-generation", model="google/flan-t5-small")
prompt = """Translate English to French:
English: Hello, how are you?
French: Bonjour, comment ça va?
English: What is your name?
French:"""
print(generator(prompt))    #Few-shot prompting
```

```python
from transformers import pipeline

generator = pipeline("text2text-generation", model="google/flan-t5-small")

prompt = "Solve this mathematical problem step-by-step: One pen costs 1 dollar. How many dollars does 6 pens costs"

print(generator(prompt))    #Chain-thought prompting
```
